## Obtención de base de datos

La siguiente base de datos fue proporcionada directamente por la empresa, la cual contiene todas las órdenes de servicio desde el 30 de septiembre de 2019 hasta finales de septiembre del presente año. Cada orden incluye diferente tipo de información, la cual se menciona a continuación:

- **No. de orden:** número identificador de la orden
- **No. de serie:** número de serie del equipo del cual es la orden
- **Modelo:** modelo del equipo
- **Num. Cliente:** número identificador del cliente
- **Fecha recepción:** fecha en la que cayó la orden de servicio
- **Estatus:** estatus de la orden (si se resolvió o no)
- **Categoría:** qué tipo de servicio se va a proporcionar
- **Ruta:** Estado de la República Mexicana a la cual pertenece
- **Técnico de visita:** nombre del ingeniero de servicio (IS) que atendió la orden de servicio
- **No. de visitas:** visitas que realizó el igeniero de servicio (IS) antes de ser resuelta la orden de servicio
- **Fecha cierre:** fecha en la que se resolvió la orden de servicio
- **T. transcurrido total:** tiempo en total que transcurrió desde que cayó la orden hasta que se cerró
- **Tiempo frente al cliente:** tiempo de mano de obra
- **T. respuesta:** tiempo que transcurrió desde que cayó la orden hasta que el ingeniero de servicio (IS) llegó a atenderla
- **Límite T. respuesta:** tiempo máximo acordado que debe tardar en llegar el ingeniero de servicio (IS) a atender la orden
- **TST:** tiempo de solución total para EXEL PITSS 
- **Límite TST:** límite de tiempo de solución total para EXEL PITSS
- **TR OK:** 0 si no se resolvió dentro del límite del tiempo, y 1 si sí se resolvió dentro del límite de tiempo
- **Alcance TR:** es el resultado: 2-TR/TR límite
- **Alcance TST:** es el resultado: 2-TST/TST límite
- **Cerrados en primera visita:** 1 si la orden fue cerrada en la primera visita, 0 si no.

```python
import pandas as pd
import numpy as np
from matplotlib import cm
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
```

```python
data = pd.read_csv('data/ordenes_servicio.csv', encoding='latin1')
data.head(10)
```

```python
tiempoT= pd.DataFrame(data,columns=['T. transcurrido total'])
tiempoC= pd.DataFrame(data,columns=['Tiempo frente al cliente'])
tiempoR= pd.DataFrame(data,columns=['T. respuesta'])
TST= pd.DataFrame(data,columns=['TST'])
limiteR= pd.DataFrame(data,columns=['Límite T. respuesta'])
limiteS= pd.DataFrame(data,columns=['Límite TST'])
TROK= pd.DataFrame(data,columns=['TR OK'])
cerrados= pd.DataFrame(data,columns=['Cerrados en primera visita'])
Tiempos=tiempoT.join(tiempoC).join(tiempoR).join(cerrados).join(TST).join(limiteS).join(limiteR).join(TROK)
Tiempos.head(80)
```

## Estandarización de datos

```python
Tiempos=tiempoT.join(tiempoC).join(tiempoR).join(TST)
Tiempos.head(80)
TST= pd.DataFrame(data,columns=['TST'])
promedio = Tiempos.mean()
des = Tiempos.std()
data_estd = (Tiempos - promedio)/des
data_estd
```

```python
plt.figure(figsize=(16,8))
plt.subplot(131)
plt.scatter(np.arange(0,len(tiempoT)),tiempoT,color='dodgerblue')
plt.title('Tiempo total transcurrido',size=18)
plt.ylabel('Tiempo en horas')
plt.subplot(132)
plt.ylabel('Tiempo en horas')
plt.scatter(np.arange(0,len(tiempoC)),tiempoC,color='darkblue')
plt.title('Tiempo frente al cliente',size=18)
plt.subplot(133)
plt.scatter(np.arange(0,len(tiempoR)),tiempoR,color='cornflowerblue')
plt.title('Tiempo de respuesta',size=18)
plt.ylabel('Tiempo en horas')
```

## Visualización de nodos y variables

$\color{maroon}{\text{1. Tiempos de mano de obra e ingeniero de servicio}}$

```python
IS = data.groupby(['Técnico de visita']).mean()
prom_TC = IS['Tiempo frente al cliente']
```

```python
plt.figure(figsize=(19,9))
plt.plot(prom_TC,'k.')
plt.ylabel('Promedio de horas frente al cliente',size=15)
plt.xlabel('Técnico o IS',size=15)
plt.xticks(rotation=90)
plt.title('Tiempos de mano de obra según el IS',size=22,color='maroon')
plt.show()
```

$\color{indigo}{\text{2. Tiempos de tiempo de respuesta y ruta de servicio}}$

```python
RS = data.groupby(['Ruta']).mean()
prom_TR = RS['T. respuesta']
```

```python
plt.figure(figsize=(19,9))
plt.plot(prom_TR,'k.')
plt.ylabel('Promedio de horas de tiempo de respuesta',size=15)
plt.xlabel('Ruta de servicio',size=15)
plt.xticks(rotation=90)
plt.title('Tiempos de respuesta según la ruta de servicio',size=22,color='indigo')
plt.show()
```

$\color{chocolate}{\text{3. Tiempos totales de solución y número de visitas}}$

```python
NV = data.groupby(['No. visitas']).mean()
prom_TT = NV['T. transcurrido total']
```
